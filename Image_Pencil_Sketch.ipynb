#### THE AIM OF THIS PROJECT IS TO CONVERT RGB IMAGE INTO THE IMAGE IN GRAYSCALE

#### IMPORTING THE LIBRARIES AND READING THE DATASET

```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("D:\LETSGROWMORE\Elon_Musk.jpg",0)
```

```python
print('Minimum color range',img.min())
print('Maximum color range',img.max())
```

#### CONVERTING IMAGE INTO GRAY SCALE

```python
plt.imshow(img,cmap='gray')
plt.title('Black and white imege')
plt.show()
```

#### INVERTING THE IMAGE

```python
inv = 255 - img
plt.title('INVERTED IMAGE')
plt.imshow(inv,cmap='gray')
plt.show()
```

#### APPLYING THE GAUSSIAN BLUR IN ORDER TO SMOOTH THE IMAGE

```python
gblur = cv2.GaussianBlur(inv,ksize=(21,21),sigmaX=0,sigmaY=0)
plt.title('Image after Gaussian blur is applied')
plt.imshow(gblur,cmap='gray')
plt.show()
```

#### DODGING AND BURNING THE IMAGE

```python
dodge= lambda image,mask: cv2.divide(image,255-mask,scale=256)
blended=dodge(img,gblur)
plt.title('THE PENCIL SKETCH OF THE IMAGE')
plt.imshow(blended,cmap='gray')
plt.show()
```
